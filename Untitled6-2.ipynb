```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import norm

df=pd.read_csv("city-facilities .csv")
df
```

```python
#data cleaning
#dropping columns geom as it is not required
df=df.drop(columns=['geom'])
#dropping columns which have null values more than 90% of the data
df=df.dropna(axis=1,thresh=0.1*len(df))
```

```python
#removing outliers
#there are places outside san francisco, these places need to be deleted as it is irrelevant
var=pd.value_counts(df.city)/len(df)
var.plot.bar()
plt.ylabel("Percent",fontsize=15)
plt.xlabel("City",fontsize=15)
plt.title("Percentage of facilities in different Cities ")
plt.legend()
```

```python
#outliers can be clearly seen using latitude and longitude
plt.scatter(df['latitude'],df['longitude'],alpha=0.5)
plt.xlabel("Latitude",fontsize=15)
plt.ylabel("Longitude",fontsize=15)
plt.title("Before removing outliers")
```

```python
#extracting the rows that have the 'city' San Francisco 
df=df[df.city.str.contains("San Francisco")]

plt.scatter(df['latitude'],df['longitude'],alpha=0.5)
plt.xlabel("Latitude",fontsize=15)
plt.ylabel("Longitude",fontsize=15)
plt.title("After removing outliers")
ax=plt.gca()
ax.set_ylim([-122.5,-120.0])
ax.set_xlim([37.3,37.9])
#thus outliers are removed
```

```python
#replacing missing values
#categorical nan values are usually replaced using previous values
#here, block_lot is independent MAR(missing at random),it cannot replaced by any value
df['block_lot']=df['block_lot'].replace(np.nan,'-')
#owned_leased nan can be replaced by previous values
df[["owned_leased"]]=df[["owned_leased"]].fillna(method='ffill')

#replacing numerical data
#numerical data is either replaced by mean,median or mode
sns.boxplot(df['gross_sq_ft'])
#since there are many outliers mean cannot be used to replace the missing values
```

```python
#thus median is used
df['gross_sq_ft']=df['gross_sq_ft'].replace(np.nan,0.0)
median=np.median(df['gross_sq_ft'])
df['gross_sq_ft']=df['gross_sq_ft'].replace(0.0,median)
```

```python
#using mode to replace the nan values in supervisor district 
from scipy import stats
m=stats.mode(df['supervisor_district'])
df['supervisor_district']=df['supervisor_district'].replace(np.nan,int(m[0]))
```

```python
zipvar=list(filter(lambda x:len(str(x))!=5,df['zip_code']))
df['zip_code']=df['zip_code'].replace(zipvar,0.0)
df=df[df['zip_code']!=0.0]
```

```python
#normalization of all numerical columns
dataf=df.copy(deep=True)
dataf["facility_id"]=((dataf["facility_id"]-dataf["facility_id"].min())/(dataf["facility_id"].max()-dataf["facility_id"].min()))
dataf["zip_code"]=((dataf["zip_code"]-dataf["zip_code"].min())/(dataf["zip_code"].max()-dataf["zip_code"].min()))
dataf["dept_id"]=((dataf["dept_id"]-dataf["dept_id"].min())/(dataf["dept_id"].max()-dataf["dept_id"].min()))
dataf["gross_sq_ft"]=((dataf["gross_sq_ft"]-dataf["gross_sq_ft"].min())/(dataf["gross_sq_ft"].max()-dataf["gross_sq_ft"].min()))
dataf["longitude"]=((dataf["longitude"]-dataf["longitude"].min())/(dataf["longitude"].max()-dataf["longitude"].min()))
dataf["latitude"]=((dataf["latitude"]-dataf["latitude"].min())/(dataf["latitude"].max()-dataf["latitude"].min()))
dataf["supervisor_district"]=((dataf["supervisor_district"]-dataf["supervisor_district"].min())/(dataf["supervisor_district"].max()-dataf["supervisor_district"].min()))
dataf
```

```python
#standardization of all the numerical columns
dataf1=dataf.copy(deep=True)
dataf1["facility_id"]=(dataf["facility_id"]-np.mean(dataf["facility_id"]))/np.std(dataf["facility_id"])
dataf1["zip_code"]=(dataf["zip_code"]-np.mean(dataf["zip_code"]))/np.std(dataf["zip_code"])
dataf1["dept_id"]=(dataf["dept_id"]-np.mean(dataf["dept_id"]))/np.std(dataf["dept_id"])
dataf1["gross_sq_ft"]=(dataf["gross_sq_ft"]-np.mean(dataf["gross_sq_ft"]))/np.std(dataf["gross_sq_ft"])
dataf1["longitude"]=(dataf["longitude"]-np.mean(dataf["longitude"]))/np.std(dataf["longitude"])
dataf1["latitude"]=(dataf["latitude"]-np.mean(dataf["latitude"]))/np.std(dataf["latitude"])
dataf1["supervisor_district"]=(dataf["supervisor_district"]-np.mean(dataf["supervisor_district"]))/np.std(dataf["supervisor_district"])
dataf1
```

```python
import math
#Hypothesis testing for gross_sq_ft
pop_mean=df["gross_sq_ft"].mean() #population mean
sample_size=1000
sample=df.sample(sample_size)
sample_mean=sample["gross_sq_ft"].mean() #sample mean
std_dev=sample["gross_sq_ft"].std() #sample standard deviation
alpha=0.05

actual_z = abs(norm.ppf(alpha/2))
hypo_z = (sample_mean - pop_mean) / (std_dev/math.sqrt(sample_size))
```

```python
print("Sample mean is ",sample_mean)
print("Sample standard deviation is ",std_dev)
print("Population mean is",pop_mean)
print("Null hypothesis is mean equal to ",pop_mean)
print("Alternate hypothesis is mean not equal to ",pop_mean)
print("alpha value is 0.05")

print('actual z value :', actual_z)
print('hypothesis z value :', hypo_z, '\n')
if hypo_z >= actual_z or hypo_z <= -(actual_z):
    print("Reject null hypothesis")
else:
    print("Failed to reject null hypothesis")
```
